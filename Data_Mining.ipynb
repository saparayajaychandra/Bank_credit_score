```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
d1 = pd.read_csv('cust_accnt.csv')
print(d1.shape)
d1.head(4)
```

```python
d2 = pd.read_csv('cust_enq.csv')
print(d2.shape)
d2.head(4)
```

```python
d2.columns
```

```python
df1 = d2.loc[:,['customer_no', 'enq_amt']]
print(df1.shape)
df1.head()
```

```python
cust_group1 = df1.groupby(['customer_no'])
print(cust_group1.sum().head())
```

```python
print(cust_group1.sum().shape)
```

```python
df_amt = pd.DataFrame(cust_group1.sum()).reset_index()
df_amt.columns = ['customer_no', 'total_enq_amt']
print(df_amt.shape)
df_amt.head(3)
```

```python
df_count = pd.DataFrame(cust_group1.count()).reset_index()
df_count.columns = ['customer_no', 'total_enq_amt']
print(df_count.shape)
df_count.head()
```

```python
df_final = pd.merge(df_amt, df_count, on = 'customer_no', how = 'inner')
print(df_final.shape)
df_final.head(3)
```

```python
df_final.to_csv('2_Cust_enquiry_final.csv', index=False)
```
